```python
# ===== Audiogram object + plotting demo =====

import importlib
import matplotlib.pyplot as plt

# Reload in case you're iterating
import audiogram_object.audiogram as ag
import audiogram_object.symbols as sym
import audiogram_object.plot_mpl as pm

importlib.reload(ag)
importlib.reload(sym)
importlib.reload(pm)

# ----- Create example audiograms -----

left_data = {
    250: 10,
    500: 15,
    1000: 30,
    2000: 40,
    4000: 50,
    8000: 65,
}

right_data = {
    250: 15,
    500: 20,
    1000: 25,
    2000: 35,
    4000: 55,
    8000: 60,
}

aL = ag.Audiogram.from_dict(left_data)
aR = ag.Audiogram.from_dict(right_data)

binaural = ag.BinauralAudiogram(left=aL, right=aR)

print("Left thresholds:", aL.to_dict())
print("Right thresholds:", aR.to_dict())

# ----- Single-ear plot -----

print("\nSingle-ear plot (Left ear)")
aL.plot(
    ear="left",
    style="ansi",
    title="Single Ear Demo",
    nr_freqs=[8000],
    masked_freqs=[1000],
)

# ----- Two-panel clinic-style plot -----

print("\nTwo-panel clinic plot")
binaural.plot(
    two_panel=True,
    style="ansi",
    title="Clinic Style Audiogram",
    left_nr=[8000],
    right_nr=[4000],
    left_masked=[1000],
    right_masked=[2000],
)

# ----- Overlay plot -----

print("\nOverlay plot")
binaural.plot(
    two_panel=False,
    style="ansi",
    title="Overlay Audiogram",
)
```
